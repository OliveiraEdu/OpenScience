# Open Science Platform

The Open Science Platform is a comprehensive platform designed to support open science research and collaboration. It provides secure ownership management, decentralized storage, search functionality, and metadata management capabilities.

## Objectives

- Support Open Science Research: The platform aims to facilitate open science research by providing a secure, transparent, and decentralized environment for collaboration and data sharing.

- Enable Secure Ownership: Blockchain technology ensures secure ownership management, enabling researchers to maintain control over their projects and data.

- Foster Collaboration: The platform's features and functions are designed to promote collaboration among researchers, facilitating the discovery and reuse of research outputs.


## Features

- Decentralized Storage: IPFS provides decentralized storage for project data.

- Blockchain-based Ownership: Blockchain technology ensures secure and transparent ownership management.

- Search Functionality: The search engine allows users to query the index for files matching specific keywords.

- Metadata Management: Project metadata is stored and managed within the platform.


## Functions

- User Management: Users can enroll/login, register projects, and have assigned roles.

- Project Creation and Management: Users can create and manage projects, which are represented by project metadata, files, owners, and IDs.

- Search Engine: The platform includes a search engine that allows users to input keywords and retrieve relevant results (files matching the keyword).

- Blockchain Integration: Projects are instantiated as accounts on a blockchain, enabling secure and transparent ownership management.

- IPFS Storage: Project metadata, files, and file metadata are stored in InterPlanetary File System (IPFS), providing decentralized and content-addressed storage.

- JSON Extraction: Files are sent to a parser for metadata extraction, which is then used to update project account attributes.

- Indexing: Indexed documents are generated from extracted metadata and stored in an index for efficient querying.

--- 

## Main Definitions


Platform: The core component of the Open Science Platform, responsible for user authentication, project creation, and data management.

User: An individual who can enroll/login to the platform, register projects, and access project data.

Project: A container for storing and managing research data, including metadata, files, and owners.

Role: A defined set of permissions and responsibilities assigned to a user within a project.

Metadata: Additional information about a project, such as title, description, keywords, and authors.

File: A digital file uploaded to a project for storage and sharing.

Project Owner: The individual responsible for managing a project's metadata, files, and access permissions.

Search Engine: A component that enables keyword-based search functionality within the platform.

---

## Components


### Open Science Platform


Enrollment/Login: User self enrollment, authentication and authorization process.

Project Registration: User can create new projects.

Role Assignment: Project Owner can assign users to specific roles within a project, determining their access permissions.

Metadata Storage: Projects store metadata for easy searching and organization.

File Upload: Researchers can upload files to their projects for storage and sharing.


### Blockchain


User Account Creation: Upon enrollment a users account is created on the blockchain for the user.

Project Account Creation: Upong a project creatiion, a corresponding project account is created on the blockchain for each project.

User Account attributes: Each user account holds attributes related to the user identification such as full name, email, ORCID, affiliation and related projects linked to the user.

Project Account attributes: Each project account holds attributes related to the project, such as project owner and CIDs for project metadata, file and file metadata.


### IPFS 


File Storage: Files are stored on the InterPlanetary File System (IPFS) for decentralized access.

JSON Metadata Storage: Project metadata is stored on IPFS, ensuring availability.

CID Generation: Each file and JSON metadata stored receives a Content-Addressable Network (CID).


### JSON Extractor


File Parsing: Files are parsed for metadata extraction. Parsed metadata JSON are sent to IPFS for storage and are indexed for search functionality.


### Indexer 


Indexed Document Generation: An index of files metadata is generated for searching.

Keyword-Based Search: Users can query the index using keywords to retrieve relevant files.

Result Display: The indexer displays a list of matching files based on the search query.

   The blockchain-based ownership management feature ensures that they retain control over their data.
